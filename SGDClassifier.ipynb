```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
from elpv_dataset.utils.elpv_reader import load_dataset
images, proba, types = load_dataset()
```

```python
index = np.arange(images.shape[0])
np.random.shuffle(index)
images = images[index]
proba = proba[index]
types = types[index]
```

```python
data_fraction = 3/4
split = int(len(images)*(data_fraction))

train_images = images[:split]
train_proba = proba[:split]
train_types = types[:split]

test_images = images[split:]
test_proba = proba[split:]
test_types = types[split:]
```

```python
from sklearn.linear_model import SGDClassifier

sgd = SGDClassifier(loss='log_loss', n_jobs=-1, n_iter_no_change=1000, max_iter=3000)
# sgd = SGDClassifier(n_jobs=-1)

train_images_flatten = train_images.reshape(np.shape(train_images)[0],-1)
test_images_flatten = test_images.reshape(np.shape(test_images)[0],-1)

sgd.fit(train_images_flatten,np.multiply(train_proba,3))
```

```python
from sklearn import metrics

sgd_pred = sgd.predict(test_images_flatten)
print(metrics.classification_report(np.multiply(test_proba,3), sgd_pred))
disp = metrics.ConfusionMatrixDisplay.from_predictions(np.multiply(test_proba,3), sgd_pred)
disp.figure_.suptitle("Confusion Matrix")
plt.show()
```

```python
#MONO CELLS ONLY

sgd_pred = sgd.predict(test_images_flatten[test_types == 'mono'])

print(metrics.classification_report(np.multiply(test_proba,3)[test_types == 'mono'], sgd_pred))
disp = metrics.ConfusionMatrixDisplay.from_predictions(np.multiply(test_proba,3)[test_types == 'mono'], sgd_pred)

disp.figure_.suptitle("Confusion Matrix")
plt.show()

#POLY CELLS ONLY
sgd_pred = sgd.predict(test_images_flatten[test_types == 'poly'])

print(metrics.classification_report(np.multiply(test_proba,3)[test_types == 'poly'], sgd_pred))
disp = metrics.ConfusionMatrixDisplay.from_predictions(np.multiply(test_proba,3)[test_types == 'poly'], sgd_pred)

disp.figure_.suptitle("Confusion Matrix")
plt.show()
```
